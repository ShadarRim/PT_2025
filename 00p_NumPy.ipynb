## Массивы

Главная особенность и элемент, с которым необходимо работать, в numpy - это массивы. Создаются массивы разными способами, которые мы сейчас разберём.

При этом всем элементы в array должны быть одного типа, что отличает его от классического списка (list) Python.

**Давайте сравним `numpy.array()` с обычным типом `list()`**

- Во-первых, списки Python динамически типизированы. Это означает, что каждый элемент списка может быть любого типа, и Python должен хранить информацию о типе каждого элемента. Это увеличивает объём используемой памяти и замедляет обработку данных.

- Во-вторых, списки Python не поддерживают векторизованные операции, то есть операции, которые можно применить сразу ко всем элементам списка без написания цикла. Это означает, что обработка больших списков может быть очень медленной.

**Преимущества `NumPy` над обычными списками Python:**

- Эффективное использование памяти. Массивы NumPy статически типизированы и плотно упакованы в памяти. Это означает, что они используют гораздо меньше памяти, чем списки Python.

- Поддержка векторизованных операций. Это означает, что можно выполнять операции над всеми элементами массива одновременно, без необходимости писать циклы. Это делает обработку данных гораздо быстрее.

- Встроенные функции для научных вычислений. NumPy предлагает множество функций для выполнения научных вычислений, таких как линейная алгебра, статистические операции и многое другое.

С точки зрения кода, мы посмотрим на разницу чуть позже.

### Создание по-разному заполненных массивов

`numpy.arange()` — это встроенная в библиотеку `NumPy` функция, которая возвращает объект типа ndarray, содержащий равномерно расположенные значения внутри заданного интервала.

Например, чтобы сгенерировать значения от одного до десяти, вы можете воспользоваться функцией `numpy.arange()`.

**Синтаксис:**

`numpy.arange(start, stop, step)`

Здесь `start` — начало интервала, `stop` — конец интервала, `step` — интервал между значениями (по умолчанию равен 1).

**В отличии от функции `range()` функция `numpy.arange()` может оперировать вещественным шагов (не только целым).**

````python
# аналог range для массивов
print(np.arange(5)) # start = 0, step = 1
print(np.arange(1, 6, 2))
````

````python
print(np.arange(1.6, 6, 2.5))
````

````python
np.array(range(6))
````

Функция `numpy.ones()` применяется для создания матрицы, состоящей из одних единиц. Часто данная функция используется для инициализации начальных значений, а также в разных статистических задачах.

**Синтаксис:**

`numpy.ones(shape, dtype=float, order='C')`

В этом синтаксисе shape — размерность массива, dtype — тип данных (указывается опционально, по умолчанию задан тип float64), order определяет порядок хранения массива в памяти. Order указывается опционально; по умолчанию задан параметр ‘C’, что означает построковое хранение. Можно поменять на ‘F’ и хранить по колонкам.

````python
np.ones((2,3))
````

Функция `numpy.zeros()` применяется для создания нулевой матрицы. Часто данная функция используется для инициализации начальных значений, а также в разных статистических задачах.

**Синтаксис:**

`numpy.zeros(shape, dtype=float, order='C')`

- `shape` — размерность массива в виде кортежа
- `dtype` — тип данных (указывается опционально, по умолчанию задан тип float64)
- `order` определяет порядок хранения массива в памяти. Order указывается опционально, по умолчанию задан параметр `C`, что означает построковое хранение. Можно поменять на `F` и хранить по колонкам.

````python
np.zeros((5,4))
````

````python
np.identity(4) # квадратную!
````

````python
# k - номер диагонали, заполненный единицами
np.eye(5, 4, k=2)
````

### Перебор элементов массива

````python
a = np.array([1, 4, 5], int)
````

````python
# простой перебор для одномерного случая
# перебор по элементам
for x in a:
   print(x)
````

````python
for x in range(len(a)):
   print(x)
````

````python
# простой перебор для многомерного случая срабаотает плохо, он перебирает по первой размерности
a = np.array([[1, 2], [3, 4], [5, 6]], float)
for x in a:
   print(x)
````

````python
# перебор правильным способом
for x in range(a.shape[0]):
  for y in range(a.shape[1]):
    print(a[x, y])
````

### Выбор элементов массива по условию

Очень важной особенностью массивов является то, что элементы в них можно брать на основании других массивов. Это особенно полезно при реализации свёрточных нейронных сетей.

````python
# это результат применения логической операции к многомерному массиву
a = np.array([[6, 4], [5, 9]], float)
a >= 6
````

````python
# а это результат фильтрации элементов
# обратите внимание, получился одномерный массив, содержащий только элементы, удовлетворяющие условию
a[a >= 6]
````

````python
a[np.logical_and(a > 5, a < 9)]
````

Обратите внимание, что если передать целочисленные значения, в качестве условий, то результат будет другой. Будут выбраные соответствующие элементы.

````python
a = np.array([2, 4, 6, 8], float)
b = np.array([0, 0, 1, 3, 2], int)
a[b]
````

````python
# Для выбора значений из многомерных массивов необходимо передать массивы, которые определяют индексы по каждому из направлений. Они должны быть, естественно, целочисленными.
a = np.array([[1, 4], [9, 16]], float)
b = np.array([0, 0, 1, 1, 0], int)
c = np.array([0, 1, 1, 1, 1], int)
a[b, c]
````

## Векторная и матричная математика с использованием numpy

Векторная математика в numpy - это главная причина того, что numpy стал ключевым модулем Python среди всех представленных модулей. Векторные вычисления позволяют значительно ускорить обработку численной информации.

Часто сравнивая Python c С++/C говорят том, что первый гораздо менее производителен. Но с учётом современных модулей верно следующее утверждение: хорошо написанная программа на Python будет производительнее, чем средняя программа на C/C++, хорошую программу на C/C++ написать крайне сложно.

В дополнении к арифметически операциям мы рассмотрим некоторые векторные операции. Глубокое погружение в их суть требует значительных знаний математики, которые мы будем получать по мере необходимости.

### Скалярное произведение

Для двух векторов a и b одинаковой длины скалярное произведение считается по следующей формуле:

$ a*b = \sum_{i=0}^{len(a)}  a_i*b_i $

````python
# скалярное произвдение векторов, также операция свёртки в свёрточных нейронных сетях

a = np.array([1, 2, 3], float)
b = np.array([0, 1, 1], float)
np.dot(a, b)

a @ b
````

### Произведение матриц

Произведение матриц - это особая математическая операция, которая не эквивалентна произведени соответствующих элементов матриц. О матричном произведении целесообразно говорить в рамках соответствующих разделов математики. Тем не менее, используя numpy легко получить матричное произведение.

````python
a = np.array([[0, 1], [2, 3]], float)
b = np.array([2, 3], float)
d = np.array([[1, 1, 1], [1, 1, 1], [1, 1, 1]], float)
````

````python
np.dot(b, a)
````

````python
np.dot(a, b)
````

````python
a@b
````

````python
# следите за размерностью, иначе ничего не получится
np.dot(b, d)
````

**Задача 4.**

Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением.

````python
import numpy as np

a = np.array([1, 3, 2, 4, 5])

b = np.zeros((len(a) + len(a)*3))

b[::3+1] = a

b
````

**Задача 7.**

Дан массив 16x16, посчитать сумму по блокам 4x4.

````python
m = np.array([[j for j in range(10)] for i in range(10)])
m
````

````python
m = np.array([[j for j in range(16)] for i in range(16)])

s = []

for i in range(4):
    for j in range(4):
        s.append(np.sum(m[(i * 4):(i * 4 + 4), (j * 4):(j * 4 + 4)]))

print(s)
````

**Задача 8.**

Дана матрица. Найти n наибольших значений в массиве. n вводится с клавиатуры.

**Задача 10.**

Дан вектор. Найти максимальный элемент в векторе среди элементов, перед которыми стоит нулевой.

Например для:

`x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])`

Ответ:
5

````python
x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0])
zeros = np.where(x == 0)[0]
zeros = np.delete(zeros, np.where(zeros == len(x) - 1))

print(np.max(x[zeros + 1]))
````
